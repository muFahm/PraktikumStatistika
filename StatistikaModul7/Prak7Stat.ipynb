```python
import pandas as pd
```

#Latihan1

```python
#memanggil dan menampilkan dataset
data_fahmi = pd.read_csv('titanic.csv',sep=";")
print(data_fahmi)
```

```python
#ngambil data pada kolom tertentu
data1 = data_fahmi.loc[:,['Age', 'Pclass', 'Survived']]
print(data1)
```

```python
#memvisualisasikan data titanic
data2 = data_fahmi[['Age', 'Pclass', 'Survived']]
data2.plot(title='Persebaran Data', x = 'Age', y = 'Pclass', kind = 'scatter', c = 'Survived', colormap = 'Paired')
```

```python
#memanipulasi data jumlah penumpang berdasarkan group Pclass
data3 = data_fahmi[['Name', 'Sex', 'Age', 'Pclass', 'Fare']]
penumpang = data3.groupby('Pclass')['Name']. nunique()
print('Jumlah Penumpang:\n', penumpang)
```

```python
#memfilter data penumpang yang selamat berdasarkan Pclass
data4 = data_fahmi[['Name', 'Sex', 'Age', 'Pclass', 'Fare']]
notsurvivedpassanger = data4['Pclass'].loc[data_fahmi['Survived']==0]
print('Penumpang yang tidak survive:\n', notsurvivedpassanger.value_counts())
survivedpassanger = data4['Pclass'].loc[data_fahmi['Survived']==1]
print('\nPenumpang yang survived:\n', survivedpassanger.value_counts())
```

```python
#manipulasi data jumlah penumpang berdasarkan group sex
data5 = data_fahmi[['Name', 'Sex', 'Age', 'Pclass', 'Fare']]
penumpang_sex = data5.groupby('Sex')['Name'].nunique()
print('Jumlah Penumpang:\n', penumpang_sex)
```
